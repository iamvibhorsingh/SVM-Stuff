```python
import seaborn as s
import matplotlib.pyplot as pl
```

```python
m = p.read_csv("mushrooms.csv")
```

```python
m
```

```python
m["odor"].unique()
```

```python
m["class"] = preprocessing.LabelEncoder().fit_transform(m["class"])
m["cap-shape"] = preprocessing.LabelEncoder().fit_transform(m["cap-shape"])
m["cap-surface"] = preprocessing.LabelEncoder().fit_transform(m["cap-surface"])
m["cap-color"] = preprocessing.LabelEncoder().fit_transform(m["cap-color"])
m["bruises"] = preprocessing.LabelEncoder().fit_transform(m["bruises"])
m["odor"] = preprocessing.LabelEncoder().fit_transform(m["odor"])
m["gill-attachment"] = preprocessing.LabelEncoder().fit_transform(m["gill-attachment"])
m["gill-spacing"] = preprocessing.LabelEncoder().fit_transform(m["gill-spacing"])
m["gill-size"] = preprocessing.LabelEncoder().fit_transform(m["gill-size"])
m["gill-color"] = preprocessing.LabelEncoder().fit_transform(m["gill-color"])
m["stalk-shape"] = preprocessing.LabelEncoder().fit_transform(m["stalk-shape"])
m["stalk-root"] = preprocessing.LabelEncoder().fit_transform(m["stalk-root"])
m["stalk-surface-above-ring"] = preprocessing.LabelEncoder().fit_transform(m["stalk-surface-above-ring"])
m["stalk-surface-below-ring"] = preprocessing.LabelEncoder().fit_transform(m["stalk-surface-below-ring"])
m["stalk-color-above-ring"] = preprocessing.LabelEncoder().fit_transform(m["stalk-color-above-ring"])
m["stalk-color-below-ring"] = preprocessing.LabelEncoder().fit_transform(m["stalk-color-below-ring"])
m["veil-type"] = preprocessing.LabelEncoder().fit_transform(m["veil-type"])
m["veil-color"] = preprocessing.LabelEncoder().fit_transform(m["veil-color"])
m["ring-number"] = preprocessing.LabelEncoder().fit_transform(m["ring-number"])
m["ring-type"] = preprocessing.LabelEncoder().fit_transform(m["ring-type"])
m["spore-print-color"] = preprocessing.LabelEncoder().fit_transform(m["spore-print-color"])
m["population"] = preprocessing.LabelEncoder().fit_transform(m["population"])
m["habitat"] = preprocessing.LabelEncoder().fit_transform(m["habitat"])
```

```python
from sklearn import svm,metrics,cross_validation
```

```python
x_train,x_test,y_train,y_test = cross_validation.train_test_split(m,mtype,test_size = 0.1)
```

```python
a = svm.SVC(kernel = 'linear',C = 10, gamma = 0.01)
```

```python
a.fit(x_train,y_train)
```

```python
y_pred = a.predict(x_test)
```

```python
scores = a.score(x_test,y_test)
```
