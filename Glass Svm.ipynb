```python
import pandas as p
from sklearn import svm,metrics,cross_validation
import matplotlib.pyplot as pl
import seaborn as s
```

```python
g = p.read_csv("glass.csv")
```

```python
gtype = g["Type"]
```

```python
#del g["Type"]
```

```python
x_train,x_test,y_train,y_test = cross_validation.train_test_split(g,gtype,test_size = 0.25)
```

```python
f = svm.SVC(kernel="rbf",C=100,gamma=0.05)
```

```python
f.fit(x_train,y_train)
```

```python
y_pred= f.predict(x_test)
```

```python
f.score(x_test,y_test)
```

```python
pl.scatter(y_test,y_pred)
pl.show()
```
